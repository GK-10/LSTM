```python
import numpy as np
import pandas as pd
import matplotlib.pyplot
matplotlib.pyplot.rcParams['figure.figsize'] = (20.0, 10.0)

data = pd.read_csv('DOP/5b_50A.csv')
print (data.shape)
data.head()
```

```python
features = ['Hours', 'Temperature']
data = data.groupby('Hours', as_index = False)[features].mean()
```

```python
X = data['Hours'].values
Y = data['Temperature'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

m = len(X)

numer = 0
denom = 0
for i in range(m):
        numer += (X[i] - mean_x)*(Y[i] - mean_y)
        denom += (X[i] - mean_x)**2
b1 = numer / denom
b0 = mean_y - (b1*mean_x)

print(b1,b0)
```

```python
max_x = np.max (X) 
min_x = np.min (X) 

x = np.linspace(min_x,max_x,1000)
y = b0 + b1 * x

matplotlib.pyplot.plot(x, y, color = '#58b970', label = 'Regression Line')
matplotlib.pyplot.scatter(X, Y, c = '#ef5423', label = 'Scatter Plot')

matplotlib.pyplot.xlabel('Hours')
matplotlib.pyplot.ylabel('Temperature')
matplotlib.pyplot.legend()
matplotlib.pyplot.show()
```

```python
ss_t = 0
ss_r = 0
for i in range(m):
    y_pred = b0 + b1 * X[i]
    ss_t += (Y[i] - mean_y)**2
    ss_r += (Y[i] - y_pred)**2
r2 = 1 - (ss_r/ss_t)
```
